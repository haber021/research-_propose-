```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Predicting Health Status
X = df_data_no_duplicates[['Age', 'Sex', 'RegionRes', 'ProvRes', 'HealthStatus', 'AgeGroup']]  # Features
y = df_data_no_duplicates['BarangayRes']  # Target

# Encoding categorical features
X_encoded = pd.get_dummies(X)

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Model training using SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)

# Predictions with SVC
y_pred_svc = svc_model.predict(X_test)

# Combine the actual and predicted values into a DataFrame
comparison_df_svc = pd.DataFrame({'Actual': y_test.reset_index(drop=True), 'Predicted': y_pred_svc})

# Count the occurrences of each class in both actual and predicted for SVC
actual_counts_svc = comparison_df_svc['Actual'].value_counts().sort_index()
predicted_counts_svc = comparison_df_svc['Predicted'].value_counts().sort_index()

# Convert counts to percentages for SVC
actual_percentages_svc = (actual_counts_svc / actual_counts_svc.sum()) * 100
predicted_percentages_svc = (predicted_counts_svc / predicted_counts_svc.sum()) * 100

# Create a DataFrame for plotting SVC results
percentages_df_svc = pd.DataFrame({'Actual': actual_percentages_svc, 'Predicted': predicted_percentages_svc}).sort_values(by='Actual', ascending=False).head(4)

# Plot the SVC data
percentages_df_svc.plot(kind='bar', figsize=(10, 6), color=['#1f77b4', '#ff7f0e'])

# Add titles and labels for SVC
plt.title('Comparison of Actual vs. Predicted Health Status (Percentage) using SVC')
plt.xlabel('Health Status')
plt.ylabel('Percentage')
plt.xticks(rotation=0)  # Rotate labels to be horizontal
plt.legend(title='Legend')
plt.show()

# Accuracy score for SVC
svc_accuracy = round(accuracy_score(y_test, y_pred_svc) * 100, 2)
print("The accuracy score achieved using SVC is: " + str(svc_accuracy) + " %")

# Model training using Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Predictions with Naive Bayes
y_pred_nb = nb_model.predict(X_test)

# Accuracy score for Naive Bayes
nb_accuracy = round(accuracy_score(y_test, y_pred_nb) * 100, 2)
print("The accuracy score achieved using Naive Bayes is: " + str(nb_accuracy) + " %")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Update features and target
X = df_data_no_duplicates[['Age', 'Sex', 'RegionRes']]
y = df_data_no_duplicates['HealthStatus']

# Encoding categorical features
X_encoded = pd.get_dummies(X)
y_encoded, y_encoded_labels = pd.factorize(y)  # Encode target as numeric labels

# Fill missing values
X_encoded = X_encoded.fillna(X_encoded.median())
X_encoded = X_encoded.apply(lambda x: x.fillna(x.mode()[0]) if x.dtype == 'object' else x)

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_encoded, test_size=0.2, random_state=42)

# Model training with RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_model.predict(X_test)

# Decode the numerical predictions back to the original labels
y_test_decoded = pd.Series(y_encoded_labels[y_test], index=X_test.index)  # Actual values
y_pred_decoded = pd.Series(y_encoded_labels[y_pred_rf], index=X_test.index)  # Predicted values

# Combine the actual and predicted values into a DataFrame
comparison_df = pd.DataFrame({'Actual': y_test_decoded, 'Predicted': y_pred_decoded})

# Count the occurrences of each class (city/municipality) in both actual and predicted
actual_counts = comparison_df['Actual'].value_counts(normalize=True) * 100  # Percentage
predicted_counts = comparison_df['Predicted'].value_counts(normalize=True) * 100  # Percentage

# Create a DataFrame for plotting
counts_df = pd.DataFrame({'Actual': actual_counts, 'Predicted': predicted_counts}).fillna(0)

# Plot the data
counts_df.plot(kind='bar', figsize=(14, 8), color=['#1f77b4', '#ff7f0e'])

# Add titles and labels
plt.title('Percentage Comparison of Actual vs. Predicted Cases in Health')
plt.xlabel('City/Municipality')
plt.ylabel('Percentage')
plt.xticks(rotation=90)  # Rotate labels to be vertical if there are many cities
plt.legend(title='Legend')
plt.show()

# Model training with GaussianNB
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Predictions with GaussianNB
y_pred_nb = nb_model.predict(X_test)

# Accuracy Scores
score_rf = round(accuracy_score(y_test, y_pred_rf) * 100, 2)
score_nb = round(accuracy_score(y_test, y_pred_nb) * 100, 2)

print(f"The accuracy score achieved using Random Forest is: {score_rf} %")
print(f"The accuracy score achieved using Naive Bayes is: {score_nb} %")
```
